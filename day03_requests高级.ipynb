### cookie
- cookie是存储在客户端的一组键值对
- cookie是由服务器段创建
- cookie应用的场景：
    - 免密登录 指定时长免密登录
    

- 在爬虫中处理cookie的两种方式
    - 手动处理
        - 将cookie手动封装到字典中，将字典作用到get/post方法的headers参数中
    - **自动处理**
        - Session对象
        - session对象的创建：requests.Session()
        - 对象的作用：
            - 可以和requests对象一样调用get/post发送请求，在使用session进行请求时，如果产生了cookie，则cookie会被自动存储到session对象中。
        - 在爬虫使用session处理cookie时，至少会被调用几次？
            - 两次，第一次捕获和存储cookie，第二次用携带的cookie发送请求
            
            

```python
import requests
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.116 Safari/537.36'
}
```

```python
# 爬取雪球网的新闻
url = 'https://xueqiu.com/statuses/hot/listV2.json?since_id=-1&max_id=75396&size=15'
json_data = requests.get(url=url, headers=headers).json()
json_data

```

```python
# 爬取雪球网的新闻
sess = requests.Session() # 携带着空的cookie
sess.get(url='https://xueqiu.com/', headers=headers) # 该次请求只是为了捕获cookie存储到session中，第一次
url = 'https://xueqiu.com/statuses/hot/listV2.json?since_id=-1&max_id=75396&size=15' # 第二次
json_data = sess.get(url=url, headers=headers).json()
json_data
```

### 代理操作
- 概念：代理服务器
- 代理服务器的作用
    - fiddler就是一个典型的代理服务器的抓包工具
    - 用来转发请求和响应，可以让代理服务器更换原始的ip地址
- 代理和爬虫之间的关联
    - 爬虫程序可能会在短时间内对指定的服务器发起高频的请求，服务器可以将该ip禁用
- 代理的匿名度：
    - 透明：对方服务器知道使用了代理，也知道使用的真实的ip
    - 匿名：对方知道使用了代理，但是不知道真实的ip
    - 高匿：对方不知道用了代理，也不知道真实的ip
- 代理的类型：
    - http：只能够代理http协议的请求
    - https：代理https协议的请求
- 如果获取代理服务器
    - 免费

        - 快代理
        - goubanjia

        - 代理精灵
